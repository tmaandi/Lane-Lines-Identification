## Project: **Finding Lane Lines on the Road** 
***
The main goal of the project is to identify and annotate the lane lines in a video of a road. A code pipepline of image processing is developed and tested, first on images and subsequently on videos. In brief, a raw image is first converted to grayscale, then filtered using a guassian kernel followed by subjecting it to Canny Edge Detection algorithm. The region of interest in the image (bare minumum area including the lane lines) is marked and everything outside is masked out. This image is then subjected to Hough transform to obtain line segment parameters representing the lines obtained by Canny Edge detection. The obtained line segments are then separated into two sets: left lane and right lane. The line segments with outlier slopes are rejected to obtain a smoother, uniform dataset representing each lane. Finally, these line segments are used to calculate an average, equivalent single line representing one lane. Once a statisfactory lane identification is observed, this pipeline is put together to work on videos, acting on images from individual video frames. In order to observe a smooth transition on video from one frame to another, a weighted moving average filter of window size 3 is implemented.

## Helper Functions

Here are some initial functions that will be used in the image processing

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
```

### Gaussian Filtering

```python
print("Now as we have the gray scaled image, we can proceed for Gaussian filtering, to be followed by Canny Edge detection\n")
print("Applying a Gaussian filter of kernel size: 5\n")
blurred_image = gaussian_blur(image_gray, 5)
# save for report
#r_blurred_image = np.copy(blurred_image)
#cv2.imwrite("./test_images_output/r_blurred_image.jpg", r_blurred_image);

plt.subplot(121)
plt.imshow(image_gray, cmap = 'gray')
plt.title('Unfiltered Image')
plt.subplot(122)
plt.imshow(blurred_image, cmap = 'gray')
plt.title('Gaussian-filtered Image')
```

### Canny Edge Detection

```python
"""
In Canny Edge Detection algorithm, all edges with gradients above the upper
threshold are selected, all edges with gradients below lower threshold are
rejected. The edges with gradients between lower and upper threshold, that 
are in contact with edges above the upper threshold, are retained, rest are
rejected.
"""
lower_threshold = 70
upper_threshold = 200
canny_image = canny(blurred_image, lower_threshold, upper_threshold)
# save for report
#r_canny_image = np.copy(canny_image)
#cv2.imwrite("./test_images_output/r_canny_image.jpg", r_canny_image);

plt.subplot(121)
plt.imshow(blurred_image, cmap = 'gray')
plt.title('Blurred Image')
plt.subplot(122)
plt.imshow(canny_image, cmap = 'gray')
plt.title('Canny-Detection Edges')
```

### Masking with region of interest and Hough Transform

```python
# Setting up the region of interest
vertices = np.array([[[110,540],[420,330],[545,330],[920,540]]], dtype = 'int32')
#vertices = np.array([[[110,480],[420,350],[555,350],[920,480]]], dtype = 'int32')
masked_image = region_of_interest(canny_image, vertices)
# save for report
#r_masked_image = np.copy(masked_image)
#cv2.imwrite("./test_images_output/r_masked_image.jpg", r_masked_image);

plt.subplot(221)
plt.imshow(canny_image, cmap = 'gray')
plt.title('Canny-Detection Image')
plt.subplot(222)
plt.imshow(masked_image, cmap = 'gray')
plt.title('Masked Image')
x = [vertices[0,0,0], vertices[0,1,0], vertices[0,2,0],vertices[0,3,0],vertices[0,0,0]]
y = [vertices[0,0,1], vertices[0,1,1], vertices[0,2,1],vertices[0,3,1],vertices[0,0,1]]
plt.plot(x, y, 'r--', lw=2)

"""
Masking done, now applying hough transform

minLineLength - Minimum length of line. Line segments shorter than this are rejected.
maxLineGap - Maximum allowed gap between line segments to treat them as single line.
rho : The resolution of the parameter r in pixels. We use 1 pixel.
theta: The resolution of the parameter \theta in radians. We use 1 degree (CV_PI/180)
threshold: The minimum number of intersections to detect a line

""" 
minLineLength = 200
maxLineGap = 5
rho = 1
theta = np.pi/180
threshold = 15

lines_img = cv2.HoughLinesP(masked_image,rho,theta,threshold,minLineLength,maxLineGap)
#print(lines_img)
draw_lines(image, lines_img)
# save for report
#r_hough_image = np.copy(image)
#r_hough_image = r_hough_image[...,::-1]
#cv2.imwrite("./test_images_output/r_hough_image.jpg", r_hough_image);

plt.subplot(212)
plt.imshow(image)
plt.title('Lines from Hough Transform')
```

### Extrapolation and Weighted Merge
** Below are some functions used in outlier rejection and extrapolation, followed by their usage. **

```python
def reject_outliers(data, m=1.5):
    """
    This function rejects the lines with slopes outside the standard deviation of m
    """
    without_outliers = np.array([(data[abs(calcSlope(data) - np.mean(calcSlope(data))) \
                                      < m * np.std(calcSlope(data))])]).reshape(-1,1,4)
    return without_outliers

def calcSlope(lines):
    """
    Lines are the output of Hough_Lines function. This function calculates slope 
    for each line segment contained in the input array "lines"
    """
    i = 0
    slopes = np.zeros_like(np.arange(np.shape(lines[:,0,0])[0]), dtype = float)
    for line in lines:
        for x1,y1,x2,y2 in line:
            if((x2 - x1) > 0.0001):
                slopes[i] = (y2-y1)/(x2 - x1)
                i = i + 1
            else:
                # If slope is to high, use the last calculated values instead
                if(i != 0):
                    slopes[i] = slopes[i-1]
                    i = i + 1
                else:
                    slopes[i] = 0
                        
    return slopes

def calc_extreme_points(lines, startY, endY):
    #removing outliers
    lines = reject_outliers(lines) 
    if (lines.size != 0):
        draw_lines(original_image_3, lines)
        # save for report
        #r_original_image_3 = np.copy(original_image_3)
        #r_original_image_3 = r_original_image_3[...,::-1]
        #cv2.imwrite("./test_images_output/r_no_outliers_image.jpg", r_original_image_3);
        """
        Feeding the outlier-less lines to polyfit to come up with a single line 
        representing one lane.

        """
        [slope, intercept] = np.polyfit(np.concatenate((lines[:,0,0],lines[:,0,2]),axis = 0),\
                                        np.concatenate((lines[:,0,1],lines[:,0,3]),axis = 0), 1)
        startX = (startY - intercept)/slope
        endX = (endY - intercept)/slope
    else:
        startX = 0
        endX = 0
    #print (startX,endX)
    return [int(startX), int(startY), int(endX), int(endY)]
    
def extrapolate_line(lines, startY, endY):
    slopes = calcSlope(lines)
    left_idx = (slopes < 0)
    right_idx = (slopes > 0)
    mean_left_slope = np.average(slopes[left_idx])
    #print(mean_left_slope)
    mean_right_slope = np.average(slopes[right_idx])
    #print(mean_right_slope)
    left_lines_img = lines[left_idx,:,:]
    #print(left_lines_img)
    right_lines_img = lines[right_idx,:,:]
    #print(right_lines_img)
    [start_left_X, startY, end_left_X, endY] = calc_extreme_points(left_lines_img, startY, endY)
    [start_right_X, startY, end_right_X, endY] = calc_extreme_points(right_lines_img, startY, endY)
    
    return np.array([[[start_left_X, startY, end_left_X, endY], [start_right_X, startY, end_right_X, endY]]])

startY = 320
endY = 540
extrapolated_lines = extrapolate_line(lines_img, startY, endY)
#print(extrapolated_lines)
#print(extrapolated_lines.shape)
if(((extrapolated_lines[0,0,0] !=0) and (extrapolated_lines[0,0,2] !=0)) or\
   ((extrapolated_lines[0,1,0] !=0) and (extrapolated_lines[0,1,2] !=0))):
    draw_lines(original_image_1, extrapolated_lines, thickness = 8)
    
plt.subplot(2,2,1)
plt.imshow(image)    
plt.title("Hough Lines - With Outliers")

plt.subplot(2,2,2)
plt.imshow(original_image_3)
plt.title("Hough Lines - Without Outliers")

#merging annotated image with original image for transparency
merged_image = weighted_img(original_image_1, original_image_2, 0.5, 0.5, 0)
# save for report
r_merged_image = np.copy(merged_image)
r_merged_image = r_merged_image[...,::-1]
cv2.imwrite("./test_images_output/solidWhiteCurve.jpg", r_merged_image);

plt.subplot(2,1,2)
plt.imshow(merged_image)
plt.title("Extrapolated Lines without Outliers")
```

```python
# Storing last three frames for smooth fame-to-frame blending
extrapolation_history = np.zeros([3,2,4],dtype = np.float64)

def process_image(image):
    # NOTE: The output is a color image (3 channel) for processing video below
    # The final output is the image where (transparent) lines are drawn on lanes
    global extrapolation_history
    original_image = np.copy(image)
    image_gray = grayscale(image)
    blurred_image = gaussian_blur(image_gray, 5)
    canny_image = canny(blurred_image, 70, 200)
    vertices = np.array([[[110,540],[420,330],[545,330],[920,540]]], dtype = 'int32')
    #vertices = np.array([[[110,480],[420,350],[555,350],[920,480]]], dtype = 'int32')
    masked_image = region_of_interest(canny_image, vertices)
    minLineLength = 200
    maxLineGap = 5
    lines_img = cv2.HoughLinesP(masked_image,1,np.pi/180,15,minLineLength,maxLineGap)
    startY = 320
    endY = 540
    extrapolated_lines = extrapolate_line(lines_img, startY, endY)
    extrapolation_history = np.append(extrapolation_history,extrapolated_lines, axis = 0)
    extrapolation_history = np.delete(extrapolation_history,0,0)
    #print(extrapolation_history, "\n")
    extrapolated_lines = np.average(extrapolation_history, axis = 0, weights = np.array([0.3,0.3,0.4],dtype = float)).reshape(1,2,4)
    #print(extrapolated_lines, "\n")
    if(((extrapolated_lines[0,0,0] ==0) and (extrapolated_lines[0,0,2] ==0)) or\
       ((extrapolated_lines[0,1,0] ==0) and (extrapolated_lines[0,1,2] ==0))):    
        pass
    else:
        draw_lines(image, extrapolated_lines.astype(int), thickness = 12)  
    merged_image = weighted_img(original_image, image, 0.5, 0.5, 0)
    return merged_image
```

### 2. Identify potential shortcomings with your current pipeline


One shortcoming is that this pipeline fails at the challenge video where lanes are curvy thoroughout the video. Secondly, there are intermittent shadowed areas on the road which are changing the image gradient.

Another shortcoming could be the region of interest masking, which is currently a carefully chosen quadrilateral centered at the bottom of the image but as we can see in the challenge video, the lower end of the image is taken by the car body which hampers the pipeline function. Even if this polygon is adjusted accordingly to fit this challenge video, this still may not work in a new video with a slightly different image orientation.
